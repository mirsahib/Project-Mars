### Earthquke model using auto regression process A(1)

```python
import pandas as pd
import numpy as np
from scipy.linalg import lstsq
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('earthquake.xlsx')
```

```python
data.head(10)
```

```python
quakes = data.iloc[:,1].values #extract data from 2nd column
```

```python
plt.plot(quakes)#time series of actual data
```

```python
initial_input = np.array(quakes[0:len(quakes)-1])
```

```python
matrix = np.column_stack((np.ones(99),initial_input))
```

```python
output = np.array(quakes[1:len(quakes)])
```

```python
coefficient = lstsq(matrix,output)[0]
```

```python
predicted_value = np.matmul(matrix,coefficient)
```

```python
plt.plot(quakes)
plt.plot(predicted_value)
plt.legend(['actual value','predicted value'],loc='upper left')
```

### coefficient of determination-R<sup>2</sup>

```python
sse = np.dot((output-predicted_value),(output-predicted_value))#sum of squared error
sst = np.dot((output-np.mean(output)),(output-np.mean(output)))#total sum of square
r_square = 1- (sse/sst)
print('R^2 = '+str(r_square))
```
