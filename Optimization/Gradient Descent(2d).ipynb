### Import necessary module

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

### Import dataset

```
url = "https://raw.githubusercontent.com/mirsahib/Dataset/master/cars.csv"
df = pd.read_csv(url)
```

### Sample dataset

```
df.head()
```

### Declaring X and Y label

```
y = df.values[:,1]#mpg
x = df.values[:,6]#wt
x = np.array(x,dtype=float)
y = np.array(y,dtype=float)
```

### Plot the data

```
plt.scatter(x,y)
plt.xlabel("Weight")
plt.ylabel("MPG")
```

### Cost function for linear regression:
J(theta0,theta1) = 1/2m sigma((theta0+theta1*x )-y)^2

```
def costFunction(m,t0,t1,x,y):
  return 1/2/m * sum([(t0 + t1*x[i] - y[i])**2 for i in range(m)])
```

### Testing cost function

```
costFunction(len(x),0,0,x,y)
```

### Gradient Descent algorithm


```
t0=0#initialize theta0
t1=0#initialize theta1
current_cost = costFunction(len(x),t0,t1,x,y)
prev_cost = 1000#initialize prev cost
alpha = 0.1
while prev_cost-current_cost>0.001:## repeat untill the cost different is less 0.001
  grad0 = 1.0/m * sum([(t0 + t1*x[i] - y[i]) for i in range(m)])#partial derivative wrt to theta0
  grad1 = 1.0/m * sum([(t0 + t1*x[i] - y[i])*x[i] for i in range(m)])#partial derivative wrt to theta1
  # update the theta_temp
  t0 = t0 - alpha * grad0
  t1 = t1 - alpha * grad1
  
  prev_cost = current_cost
  current_cost = costFunction(len(x),t0,t1,x,y)
```

```
print("intercept: "+str(t0)+"\nslope: "+str(t1))
```

### Straight Line equation: 
y = theta1 * x + theta0

```
y_pred = t0+t1*x
```

```
plt.scatter(x,y)
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.plot(x,y_pred,color='red')
```
